```python

import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv("Sigma Web Development Course - Web Development Tutorials in Hindi 🗿.csv")
```

```python
df = df['Title']
```

```python
indices_to_delete = np.arange(50,139).tolist()
indices_to_delete
df = df.drop(indices_to_delete)
```

```python
titles = df.tolist()
video_titles = []
for i in range(len(titles)):
    titles[i] = titles[i].split(" |")[0]
    if ("Exercise" in titles[i]) == False:
        video_titles.append(titles[i])
```

```python
videos = os.listdir("E:\DS CWH Practice\RAG Based Project\HTML_CSS_COURSE_VIDEOS")
for video_name in videos:
    print(video_name)

```

```python
for old_name in videos:
    old_path = os.path.join("E:\DS CWH Practice\RAG Based Project\HTML_CSS_COURSE_VIDEOS", old_name)
    
    tutorial_number = int(old_name.split("(",1)[1].split(")",1)[0])
    
    ext = os.path.splitext(old_name)[1]
    
    new_path = os.path.join("E:\DS CWH Practice\RAG Based Project\HTML_CSS_COURSE_VIDEOS", str(tutorial_number) + " - " + video_titles[tutorial_number] + ext)
    
    os.rename(old_path, new_path)
```
